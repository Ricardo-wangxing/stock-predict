```python
code_list = list(cursor.fetchall())
```

```python
code_list = [i[0] for i in code_list]
code_list
```

```python
sql = "insert into all_stock_price_three_year(code, time, open, close, high, low) values (%s,%s,%s,%s,%s,%s);"
for i in range(0,len(code_list),4):
    df_data_three = jq.get_price(code_list[i:i+4], start_date='2019-01-01', end_date='2022-01-01', frequency='daily', fields=['open','close','high','low'], skip_paused=True, fq='none', count=None)
    data = []
    for idx,row in df_data_three.iterrows():
        data.append((row['code'],row['time'],row['open'],row['close'],row['high'],row['low']))
    cursor.executemany(sql,data)
db.commit()
```
